### Step 1: Describe Linear Regression Models 

Answer the following questions after reading the resource material and watching the video lectures linked in the Linear Regression Theory section:

- In your own words, briefly describe a simple linear regression model.
    - A simple linear regression model is a way to estimate how one dependent variable (y) is related to another independent variable (x). 
    - y = Constant + b1 * x1 + Error
    - With the x points, we will create a best line to fit those points and predict the relationship between the independent variable and the dependent variable (ie. linear regression will show a cause and effect relationship).
- What type of machine learning is linear regression?
    - Linear regression is Supervised Machine Learning, or in other words.. We provide the model with data, it learns and then predicts outcomes/relationships based on that data for new inputs. 
- What is a “line of best fit” using the OLS method?
    - The OLS method creates a line that minimizes the overall deifference between predicted values and actual values resulting in the best approximation of the relationship between two variables. This line is considered the line of best fit because it minimizes the overall error between the prediction and actual data points. 
- What is the difference between correlation and regression?
    - The biggest difference is that correlation does not mean causation.. Thus correlation quantifies the relationship between two variable and is represented as a single point. On the other hand, regression shows causation or how one independent variable is affected by one or multiple dependent variables. Regression is represented via a best fitting line from many points. 
- Describe a scenario when you would use linear regression and why you would use it. 
    - One example where you could use linear regression is for looking/predicting home values. The y or independent variable would be the price of the home, which could be compared to several independent or x variables like square footage, number of rooms, year of construction, etc. By utilizing a linear regression model, we could then see how much cause and effect these variables have on the price of a home. 


### Step 2: Describe Linear Regression Assumptions 

In your own words, describe the following linear regression assumptions after reading the resource material and watching the video lectures linked in the Linear Regression Theory section:

- Linearity
    - All x variables should show a linear relationship with the y (dependent variable)
    - This can be tested by creating a scatter plot with all independent variables compared to the dependent variable
    - If non linear the data should either be transformed using log or exponents, run a non-linear regression, or thrown out
    - If all relationships are linear then a linear regression can be performed 
- Normality
    - The error term should be normally distributed or errors are symetrically distributed around zero and follow a bell-shape curve. If the error term does not follow normality then alternate statistical models or techniques should be utilized 
- Homoscedasticity
    - Errors across the model should have an equal variance 
    - This can be prevented by looking for omitted variable bias, deleting outliers, or performing a log transformation -> this will create a semilog model (as x increases by 1 unit y increases by b1 percent)
    - You can also create a log log model
- No multicollinearity
    - Two or more variables have a high correlation
    - It can be fixed by dropping one of the two variables or transforming them into one variable
    - Can be discovered by looking at correlations between each indepedent variables
- No endogeneity
    - Also known as omitted variable bias, it is when a relevant independent variable is not included in your model, which in turn creates biased or counterintuitive predictions. 
    - It is important to consider all independent variables before running multiple linear regression because no endogeneity is difficult to determine or fix.
- No autocorrelation
    - aka no serial correlation means that all observations are independent of each other or the value of a specific observation doesn't depend on the values of preceding or subsequent observations 
    - this can be found by looking for patterns in the data across a scatter plot or by using durbin watson stat (2 indicates no autocorrelation below one and above 3 are signs that there is autocorrelation)
    - autocorrelation cannot be relaxed meaning there are no work arounds to transform the data for a linear regression model

### Step 3: Describe How to Interpret Results from Correlation Table in Model Summary 

In your own words, describe the following terms after reading the resource material and watching the video lectures linked in the Linear Regression Theory section:

- Coefficient of Constant / Intercept (b0)
    - The Coefficient of the represents the value of the dependent variable when all the variables are 0. If we were looking at a linear regression of the value of cars dependent on age. The coefficient of constant or intercept would be the value of a car that has 0 years of use or is brand new (ie 20,000).
- Coefficient of Independent Variable (B1)
    - The coefficient of Independent Variable represents average change for the dependent variable for every one unit change in the independent variable. I like to think of it as the power of the independent variable as it influense the slope of the line. 
- Standard Error
    - Standard of Error is a method to measure how precise a prediction may be with given data like sample size and variability. A small standard error shows a precise prediction and a large standard error shows an inaccurate prediction or model. 
- T-Statistic
    - Number used to measure significance of an estimated parameter. Usually difficult to read so it is then used to create the p-value and reject or accept the null hypothesis. 
- P-Value of T-Statistic (from the independent variable)
    - The T-Statistic is utilized to calculate the P-Value. In a correlation summary table, each independent variable is given a T-Statistic and a P-Value. P-Value ranges on a scale of 0 - 1 and allows a person to see how significant the independent variable is on the dependent variable. For example if the p-value is 0.1 there is a 10% chance that is has no effect on the dependent variable. Typically, we look for a p-value less than .05 to reject the null hypothesis.


### Step 4:  Explain R-Squared and adjusted R-squared

In your own words, describe the following terms after reading the resource material and watching the video lectures linked in the Linear Regression Theory section:

- R-squared
    - R-squared = explained variability / total variability 
    - By deriving the above equation, we can see that R squared shows how well our model fits our data on a scale of 0 to 1. For example if the score is .5, this would mean 50% of the variability is caused by the independent variables.
    -  Traditionally we're looking at scores between 0.2 to 0.9 (0 means no effect, 1 would mean all variability from the independent varieables)
    - As we increase the number of independent variables the R score will increase which is why we typically use adjusted R-squared for multivariate models. 
- Adjusted R-squared
    - Simliar to r-squared in that the adjusted r-squared is used to measure how well your model fits a set of data, but penalizes for excessive use of variables (more variables would innately increase r-squared score and bias results)
    - Used for any multivariante modeling which is common because typically a dependent variable is affected by serveral dependent variables (not just one)
    - Want to throw out any meaningless data or x variables so that we do not penalize the adjusted r-squared score
